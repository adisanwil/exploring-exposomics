```python
def tanimoto_matrix(X: np.ndarray) -> np.ndarray:
    """
    X: array of shape (n_samples, n_bits), dtype bool or {0,1}
    returns: (n_samples, n_samples) array of Tanimoto coefficients
    """
    # make sure it's 0/1 ints
    X = X.astype(np.uint8)

    # intersection for all pairs: (n, n)
    # this is sum of bitwise AND since bits are 0/1
    inter = X @ X.T  # equivalent to np.dot(X, X.T)

    # bit counts per fingerprint
    bit_counts = X.sum(axis=1)  # shape (n,)

    # union for all pairs
    union = bit_counts[:, None] + bit_counts[None, :] - inter

    # avoid division by zero (e.g. all-zero fingerprints)
    with np.errstate(divide="ignore", invalid="ignore"):
        tanimoto = inter / union
        tanimoto[union == 0] = 0.0

    # self-similarity = 1
    np.fill_diagonal(tanimoto, 1.0)

    return tanimoto

## This is used for the Lingo Distance Metric

def integral_tanimoto(a, b) -> float:
  # T = (sum(l-[Na,j - Nb,j]))/l
    keys= set(a.keys()) | set(b.keys())

    if not keys:
      return 0

    num = 0
    den = 0

    for k in keys:
      aj=float(a.get(k,0))
      bj=float(b.get(k,0))
      l = aj+bj
      if l==0:
        continue
      diff = abs(aj-bj)

    #avoid divison by zero
      term = (l - diff)/l

      num += term
      den += 1


    return num/den if den > 0 else 0

def tanimoto_matrix_integral(X: list[dict]) -> np.ndarray:
    """
    X: list of n_samples with different lenght of fragments
    returns: (n_samples, n_samples) array of Tanimoto coefficients
    """

    n = len(X)
    M = np.zeros((n, n), dtype=float)


    for i in range(n):
        M[i,i] = 0
        for j in range(i+1,n):
          tc = integral_tanimoto(X[i],X[j])
          M[i,j] = tc
          M[j,i] = tc

    return M
```

```python
from scipy.spatial.distance import braycurtis

## Comparison of logarithmic bray-curtis concentrations

def bray_curtis_comparison(dim: int, input_data_t: np.ndarray, tp_val, alpha_inp: int) -> np.ndarray:
    Results_bcc = np.zeros((dim,dim))
    print("Bray-Curtis similarity is calculating")
    with warnings.catch_warnings(action="ignore"):
        for i in range(1,dim):
            if i%10 == 0:
                print(f"{i}/{dim}")
            for j in range(i):
                corr = braycurtis(np.log2(np.array(input_data_t[:, i], dtype=float)+1), np.log(np.array(input_data_t[:, j], dtype=float)+1))
                Results_bcc[i,j] = corr

        np.save(f"{OUTPUT_DIR}Bray-Curtis_{int(tp_val)}", Results_bcc)
        print(f"Bray-Curtis similarity is saved")
    return Results_bcc

def dice_sorensen_comparison(dim: int, input_data_t: np.ndarray, tp_val, alpha_inp: int):
    Results_dsc = np.zeros((dim,dim))
    print("Dice-Sorensen similarity is calculating")
    with warnings.catch_warnings(action="ignore"):
        for i in range(1,dim):
            if i%10 == 0:
                print(f"{i}/{dim}")
            for j in range(i):
                ## Convert concentration data to just presence/absence data
                ## i.e. convert to binary
                binary_input_i = (input_data_t[:, i] != 0).astype(int)
                binary_input_j = (input_data_t[:, j] != 0).astype(int)
                corr = braycurtis(np.array(binary_input_i, dtype=float), np.array(binary_input_j, dtype=float))
                # if p_value < alpha_inp:
                ## Here we store the dissimilarity, not distance
                Results_dsc[i,j] = corr
        np.save(f"{OUTPUT_DIR}Dice-Sorensen_{int(tp_val)}", Results_dsc)
        print(f"Dice-Sorensen similarity is saved")
    return Results_dsc

## With p-value

# def bray_curtis_comparison(dim: int, input_data_t: np.ndarray, tp_val, alpha_inp: int) -> np.ndarray:
#     Results_bcc = np.zeros((dim,dim))
#     with warnings.catch_warnings(action="ignore"):
#         for i in range(1,dim):
#             for j in range(i):
#                 print(type(input_data_t[:, i]))
#                 corr, p_value = braycurtis(np.log(np.array(input_data_t[:, i], dtype=float)), np.log(np.array(input_data_t[:, j], dtype=float)))
#                 corr = abs(corr)
#                 if p_value < alpha_inp:
#                     Results_bcc[i,j] = corr
#         np.save(f"{OUTPUT_DIR}'Bray-Curtis'_{int(tp_val)}", Results_bcc)
#         print(f"Bray-Curtis is saved")
#     return Results_bcc
```

```python
from scipy.spatial import distance

def canberra_distance(dim_input, data_input, tp_input):
    print("Canberra distance is calculating")
    Results_canberra = np.zeros((dim_input, dim_input))
    with warnings.catch_warnings(action="ignore"):
        for i in range(1,dim_input):
            for j in range(i):
                corr = distance.canberra(data_input[:, i]+1, data_input[:, j]+1)
                corr = abs(corr)
                Results_canberra[i,j] = corr

    np.save(f"{OUTPUT_DIR}Canberra distance_{int(tp_input)}", Results_canberra)
    print("Canberra distance is saved")

    return Results_canberra

## This is Hamming distance, i.e binary canberra distance, not implemented (but might be later)

# def canberra_distance_binary(dim_input, data_input, tp_input):
#     print("Canberra distance binary is calculating")
#     Results_canberra = np.zeros((dim_input, dim_input))
#     with warnings.catch_warnings(action="ignore"):
#         for i in range(1,dim_input):
#             for j in range(i):
#                 corr = distance.canberra(data_input[:, i]+1, data_input[:, j]+1)
#                 corr = abs(corr)
#                 Results_canberra[i,j] = corr

#     np.save(f"{OUTPUT_DIR}Canberra distance binary _{int(tp_input)}", Results_canberra)
#     print("Canberra distance binary is saved")

#     return Results_canberra
```
